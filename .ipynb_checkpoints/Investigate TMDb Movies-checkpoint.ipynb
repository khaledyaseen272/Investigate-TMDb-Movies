### Data Cleaning

```python
df_clean = df.copy()
```

```python
# Id is string .
df_clean.id = df_clean.id.astype('str')

# release_year is string .
df_clean.release_year = df_clean.release_year.astype('str')

# release_date is datetime .
df_clean.release_date = pd.to_datetime(df_clean.release_date)
```

```python
# convert bedget , revenue , budget_adj and revenue_adj which equal to 0 and wierd values less than 1000 to be NAN .
df_clean.loc[df_clean['budget'] < 1000, 'budget'] = np.nan
df_clean.loc[df_clean['revenue'] < 1000, 'revenue'] = np.nan
df_clean.loc[df_clean['budget_adj'] < 1000, 'budget_adj'] = np.nan
df_clean.loc[df_clean['revenue_adj'] < 1000, 'revenue_adj'] = np.nan
```

```python
# test 1 
df_clean.info()
```

## Exploratory Data Analysis

### Research Question 1 (for whole time : with high budget , will the movie succed and get high revenue ?)

```python
budget_revenue = df_clean.groupby('budget_adj').revenue_adj.mean().sort_values(ascending= False).head(20)
budget_revenue 
```

```python
x = budget_revenue.index
y = budget_revenue.values
```

```python
plt.plot(x, y, 'ro');
plt.xlabel('Budget_adj');
plt.ylabel('Revenue_adj');
plt.title('Relation between the highest revenues for movies relevant budget',loc='left');
plt.show();
```

### Finding 1 : movie success don't depend only on budget but there are many other factors .

### Research Question 2  ( Do the vote count depend on revenue ?)

```python
vote_count_revenue = df_clean.groupby('revenue_adj').vote_count.max().sort_values(ascending= False).head(20)
vote_count_revenue  
```

```python
x1 = vote_count_revenue.index
y1 = vote_count_revenue.values
x1, y1
```

```python
slope, intercept, r, p, stderr = scipy.stats.linregress(x1, y1)
line = f'Regression line: y={intercept:.2f}+{slope:.2f}x, r={r:.2f}'
```

```python
fig, ax = plt.subplots()
ax.plot(x1, y1, linewidth=0, marker='s', label='Data points')
ax.plot(x1, intercept + slope * x1, label='Relation between vote counts and revenues')
ax.set_xlabel('Revenues')
ax.set_ylabel('Vote Counts')
ax.legend(facecolor='white')
plt.show()
```

### Finding 2 : Vote counts depends on movie revenues with positive correlation .

### Research Question 3  ( what the highest genres in every year ?)
